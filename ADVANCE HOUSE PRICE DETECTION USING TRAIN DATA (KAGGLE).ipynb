## Missing Values Treatment


```python
features_nan =[features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtypes=='O']

for features in features_nan:
    print("{} :{}% missing values".format(features,np.round(dataset[features].isnull().mean(),4)))
```

```python
### Replacing missing values with new variable#
def replace_cat_features(dataset, features_nan):
    data=dataset.copy()
    data[features_nan]= data[features_nan].fillna('Missing')
    return data

dataset= replace_cat_features(dataset,features_nan)
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtypes!='O']


for features in numerical_with_nan:
    print("{}: {}% missing value".format(features,np.around(dataset[features].isnull().mean(),4)))
```

```python
## Replacing Missing value
## Lets replace missing values using median as there outliers#
for features in numerical_with_nan:
    median_values = dataset[features].median()
    
    ##creating new feature to capture nan values#
    dataset[features+'nan']=np.where(dataset[features].isnull(),1,0)
    dataset[features].fillna(median_values, inplace= True)
    
dataset[numerical_with_nan].isnull().sum()    
    
    
```

## Handling Rare Categorical features 

#### In this feature we remove categorical feature which is less than 1% of observation

```python
categorical_features= [features for features in dataset.columns if dataset[features].dtypes=='O']
categorical_features
```

```python
for features in categorical_features:
    temp = dataset.groupby(features)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[features]= np.where(dataset[features].isin(temp_df), dataset[features], 'Rar Var')
```

```python
dataset.head(100)
```

#### Labeling Categorical features

```python
for features in categorical_features:
    labels_ordered = dataset.groupby([features])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i , k in enumerate(labels_ordered,0)}
    dataset[features]=dataset[features].map(labels_ordered)
    
```

# FEATURE SELECTION

```python
##Importing libraries #
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#for feature selection #
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
# to visualize all dataframe in columns#
pd.pandas.set_option('display.max_columns', None)
```

```python
##loading dataset#
dataset =pd.read_csv('C:\\Users\\Rakesh\\X_train.csv')
```

```python
##Capturing dependent feature#
y_train=dataset[['SalePrice']]
```

```python

## drop dependent feature from dataset
X_train=dataset.drop(['Id','SalePrice'],axis=1)
```

 1.Now Let apply feature selection  
 2.Applying Lasso Regression Model  
 3.selction of alpha value  
 4.Then selecting SelectFromModel to select features which co efficient are non-zero

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
##Let print total &selected features#
selected_feat = X_train.columns[(feature_sel_model.get_support())]

print('total features:{}'.format((X_train.shape[1])))
print('selected features:{}'.format(len(selected_feat)))
print('features with co-eff shrunk to zero:{}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```
